```python
import pandas as pd
df = pd.read_csv("./hackathon_song_normalize.csv")
```

```python
# First, you split the strings into lists
df['genre'] = df['genre'].str.split(',')

# Then you can explode the lists into separate rows
df = df.explode('genre')
df['genre'] = df['genre'].apply(lambda x : x.lower().replace(" ",""))
df['duration'] = df['duration_ms']/1000
```

```python
df[[ "year", "genre", "duration", "popularity"]].to_csv("songs_genre_exploded.csv")
```
